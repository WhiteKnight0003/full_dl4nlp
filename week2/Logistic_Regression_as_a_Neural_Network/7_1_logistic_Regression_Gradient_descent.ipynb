# Sau khi vectorzation cho thut ton trn - ta khng cn for na

```python
#  M Python y  (vectorized)
import numpy as np

# J =0, db =0
# dw = np.zeros((n))

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def propagate(X, y, w, b):
    m = X.shape[1]  # s mu hun luyn (m).
    
    # Forward
    Z = np.dot(w.T, X) + b #  w.T chuyn v (transpose) - chuyn w t ma trn (nx1) thnh (1xn) -> Do X l (nxm) - ch c (1xn) . (nxm) = (1xm)
    A = sigmoid(Z) # A array with A[i] = Sigmoid(Z[i]) 
    # y m = A = Sigmoid(Z)  - a nh  u  ch d on cho 1 v d cn A l c 1 mng lun
    cost = - (1/m) * np.sum(y * np.log(A) + (1 - y) * np.log(1 - A)) # tnh J(w,b)
    
    # Backward
    dZ = A - y #(1xm)  - dZ chnh l vector sai s gia d on v thc t.
    dw = (1/m) * np.dot(X, dZ.T) # (nxm).(mx1) = (nx1)  - Mi phn t trong dw l o hm ca J theo tng thnh phn ca vector trng s wj 
    db = (1/m) * np.sum(dZ) #(1xm)  

    return dw, db, cost
```
