# Sau khi vectorzation cho thuật toán trên - ta không cần for nữa

```python
# 💻 Mã Python đầy đủ (vectorized)
import numpy as np

# J =0, db =0
# dw = np.zeros((n))

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def propagate(X, y, w, b):
    m = X.shape[1]  # số mẫu huấn luyện (m).
    
    # Forward
    Z = np.dot(w.T, X) + b #  w.T chuyển vị (transpose) - chuyển w từ ma trận (nx1) thành (1xn) -> Do X là (nxm) - chỉ có (1xn) . (nxm) = (1xm)
    A = sigmoid(Z) # A array with A[i] = Sigmoid(Z[i]) 
    # y mũ = A = Sigmoid(Z)  - a nhỏ ở đầu đề chỉ dự đoán cho 1 ví dụ còn A là cả 1 mảng luôn
    cost = - (1/m) * np.sum(y * np.log(A) + (1 - y) * np.log(1 - A)) # tính J(w,b)
    
    # Backward
    dZ = A - y #(1xm)  - dZ chính là vector sai số giữa dự đoán và thực tế.
    dw = (1/m) * np.dot(X, dZ.T) # (nxm).(mx1) = (nx1)  - Mỗi phần tử trong dw là đạo hàm của J theo từng thành phần của vector trọng số wj 
    db = (1/m) * np.sum(dZ) #(1xm)  

    return dw, db, cost
```
