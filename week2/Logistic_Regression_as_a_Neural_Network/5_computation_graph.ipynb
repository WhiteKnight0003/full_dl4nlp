- Cch tnh ton vi computation graph l thc hin mt chui cc php ton n gin theo th t nh sn bi cc mi tn, gi l lan truyn tin (forward propagation).

- V c, n c u tin php ton. Tuy nhin, s u tin ny khng da vo quy tc bn phi nh (nh "nhn chia trc, cng tr sau"), m c quyt nh bi chnh cu trc ca  th.

- Cch tnh ton (Lan truyn tin)
    - Bn ch cn i theo cc mi tn t u vo n u ra. Vi v d J=3(a+bc) v cc gi tr a=5, b=3, c=2:
    - Tnh nt u = bc:
        +  th cho thy u cn u vo l b v c.
        + Tnh ton: u = 3 * 2 = 6.
    - Tnh nt v = a + u:   
        + Nt ny cn u vo l a v kt qu t nt u. Bn khng th tnh v trc khi c kt qu ca u.
        + Tnh ton: v = 5 + 6 = 11.
    - Tnh nt cui cng J = 3v:
        + Nt ny cn u vo l kt qu t nt v.
        + Tnh ton: J = 3 * 11 = 33.

 Kt qu cui cng l 33, khp vi nhng con s trn  th.
