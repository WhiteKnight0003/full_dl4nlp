1 . Gradient Descent l g?
- Gradient Descent (Gim theo Gradient) l mt thut ton ti u ha lp i lp li  tm gi tr cc tiu ca mt hm s.

- Hy tng tng bn ang ng trn mt sn ni trong m ti v mun i xung ni thp nht (thung lng). Bn khng th nhn xa, v vy ti mi bc chn, bn s tm xem hng no dc nht xung di v i mt bc nh theo hng . Bn lp li qu trnh ny cho n khi bn khng th i xung c na, lc  bn   y. Gradient Descent hot ng y ht nh vy vi hm chi ph J(w,b).

- "Gradient" chnh l mt vector ch hng dc nht ti mt im.
- "Descent" (i xung) ngha l chng ta di chuyn ngc hng vi gradient  i n im thp hn.

2 . Ti sao phi tm w, b  J(w, b) nh nht?
- Hm chi ph J(w, b) i din cho tng sai s hay "mc  sai" ca m hnh trn ton b tp d liu.
- Mt gi tr J(w, b) cao c ngha l cc d on ca m hnh ang rt xa so vi thc t, tc l m hnh ang hot ng t.
- Mt gi tr J(w, b) thp c ngha l cc d on ca m hnh rt gn vi thc t, tc l m hnh ang hot ng tt.

- Do , mc tiu ca vic hun luyn m hnh chnh l tm ra mt cp tham s w v b khin cho tng sai s J(w,b) l nh nht c th (minimize). Khi , chng ta c mt m hnh d on chnh xc nht.

3 . Ti sao J(w, b) phi l hm li (Convex)?
- Hm chi ph J(w, b) cn phi l hm li (convex)  m bo thut ton Gradient Descent c th tm thy 1 IM CC TIU TON CC DUY NHT mt cch ng tin cy.

- Hm li (Convex Function): Ging nh mt ci bt (nh  th 3D trn slide), n ch c mt im y duy nht (cc tiu ton cc). Bt k bn bt u t u trn b mt, nu bn lun i xung, bn chc chn s n c im thp nht .

- Hm khng li (Non-convex Function): Ging nh mt dy ni gp ghnh vi nhiu thung lng (nh  th sng ~ v tay). N c nhiu "im y" khc nhau (cc tiu cc b). Nu bn bt u  mt v tr khng thun li, Gradient Descent c th a bn xung mt thung lng cn v b mc kt  , trong khi im y su nht (li gii tt nht) li nm  mt thung lng khc.

- -> Bng cch chn mt hm mt mt (loss function) ph hp  J(w,b) l hm li, chng ta m bo rng Gradient Descent s khng b "lc ng".

4 . Ti sao li chn mt im ri i xung y?
- y chnh l cch thut ton Gradient Descent hot ng mt cch trc quan:
    - Bc 1: Khi to (Chn mt im ban u)
        + Chng ta khng bit w v b tt nht l g khi bt u. V vy, thut ton s khi to chng vi cc gi tr ngu nhin. im (w, b) ngu nhin ny tng ng vi mt im bt k trn b mt ca hm chi ph J(w,b).

    - Bc 2: Lp li vic i xung
        + Ti im hin ti, thut ton tnh gradient  tc l hng dc nht i ln.
        + N s i mt bc nh theo hng ngc li vi gradient (hng dc nht i xung). Kch thc ca bc i ny c gi l tc  hc (learning rate).
        + Sau bc i , chng ta c mt cp (w, b) mi, tng ng vi mt im thp hn trn b mt.
  
    - Bc 3: Dng li
        + Qu trnh ny c lp i lp li. Sau mi bc, chng ta li tin gn hn n y. Thut ton s dng li khi n n rt gn y, ni m  dc gn nh bng 0 (khng th i xung thm c na), hoc sau mt s lng vng lp  nh trc.

    - -> im cui cng m thut ton dng li chnh l cp (w, b) ti u m chng ta cn tm.

- Gi tr x v y khng phi l th chng ta "gi s" mt cch ngu nhin. Trong thc t, x v y chnh l d liu m bn c.
- Ni mt cch n gin:

- x l u vo (input): c im m bn bit.
- y l u ra (output): Kt qu thc t tng ng vi u vo .
- -> Mc tiu ca m hnh l hc c mi quan h gia x v y t d liu bn cung cp.
