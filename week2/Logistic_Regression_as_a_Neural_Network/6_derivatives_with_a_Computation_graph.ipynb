- Chắc chắn rồi! Các đường màu đỏ và ký hiệu da = 3, dv = 3 thể hiện quá trình lan truyền ngược (backward pass) để tính đạo hàm.
- Hiểu một cách đơn giản, da, dv là ký hiệu viết tắt cho đạo hàm của kết quả cuối cùng (J) theo các biến trung gian (a, v).
    + dv = 3 thực chất có nghĩa là   dJ/dv = 3
    + da = 3 thực chất có nghĩa là   dJ/da = 3
