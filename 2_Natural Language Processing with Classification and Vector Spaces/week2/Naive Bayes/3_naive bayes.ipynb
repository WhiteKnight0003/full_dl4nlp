- Zero-Frequency Problem : trong naive bayes , các xác suất được nhân với nhau , nếu bất kì xác suất nào bằng 0 - thì toàn bộ xác xuất của cảa câu cho lớp đó = 0 -> Điều này khiến mô hình trở nên quá tuyệt đối và dễ bị sai khi gặp dữ liệu mới (ví dụ, một câu tiêu cực mới có chứa từ "because")

- Giải pháp: Để giải quyết vấn đề này, người ta thường dùng một kỹ thuật gọi là Làm mịn Laplace (Laplace Smoothing). Kỹ thuật này cộng một hằng số nhỏ (thường là 1) vào số đếm của tất cả các từ, để đảm bảo rằng không có từ nào có xác suất chính xác bằng 0
