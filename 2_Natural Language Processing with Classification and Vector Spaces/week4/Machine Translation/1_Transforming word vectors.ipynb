# 3. Giải pháp: Phép biến đổi bằng Ma trận (Transform with a Matrix)
- Đây là phần cốt lõi của sơ đồ. Nó đề xuất một giải pháp élégant: học một ma trận biến đổi (transformation matrix).
- Quá trình:
    - Lấy vector của từ gốc trong tiếng Anh: [1, 0, 1] (vector của "cat").
    - Nhân vector này với một "ma trận biến đổi" đã được "học" trước đó.
    - Kết quả của phép nhân này là một vector mới: [2, 3, 2].

- Mục tiêu: Vector mới [2, 3, 2] này giờ đây nằm trong không gian vector của tiếng Pháp. Mục tiêu của việc "học" ma trận là để vector biến đổi này gần nhất có thể với vector thực tế của từ dịch đúng.

- Trong ví dụ này, vector biến đổi [2, 3, 2] rất gần với vector thực tế của từ "chat" là [2, 3, 3]. Điều này cho thấy phép biến đổi đã thành công. Máy tính chỉ cần tìm từ có vector gần nhất với [2, 3, 2] trong kho từ vựng tiếng Pháp và nó sẽ tìm thấy "chat".

- X (Ma trận nguồn): Đúng là ma trận chứa thông tin của các từ tiếng Anh. Chính xác hơn, nó không lưu trữ chữ "cat", "dog" mà lưu trữ vector (một dãy số) đại diện cho các từ đó.

- R (Ma trận biến đổi):  Đây chính là "bộ não" mà mô hình cần phải học để thực hiện việc ánh xạ từ không gian tiếng Anh sang không gian tiếng Pháp.

- Y  (Ma trận mục tiêu): Đây là điểm cần làm rõ nhất. Y  là ma trận mục tiêu hay đáp án đúng. Nó chứa các vector của các từ tiếng Pháp tương ứng mà chúng ta muốn hướng tới.

- Đúng vậy, hình ảnh này mô tả chính xác quy trình thuật toán để "giải" hay "tìm" ra ma trận R.

- Đây là một phương pháp rất phổ biến trong lĩnh vực trí tuệ nhân tạo và học máy, được gọi là Hạ Gradient (Gradient Descent). Quá trình này diễn ra theo từng bước lặp lại để tinh chỉnh R ngày càng tốt hơn.
  
1 . Khởi tạo R (Initialize R)
- Đầu tiên, chúng ta không biết R là gì. Vì vậy, ta bắt đầu bằng cách tạo ra một ma trận R với các giá trị ngẫu nhiên. Đây là "phỏng đoán" ban đầu của chúng ta.

2 . Vòng lặp (In a loop)
- Sau khi có R ban đầu, chúng ta thực hiện một vòng lặp để liên tục cải thiện nó. Mỗi vòng lặp bao gồm 3 bước sau:

    a. Tính toán Sai số (Loss) 
        - Ý nghĩa: Công thức này đo lường "mức độ sai lệch" giữa kết quả dự đoán (XR) và kết quả đúng (Y).
        - XR - Y là ma trận chứa các sai số.
        - || ... ||_F (gọi là chuẩn Frobenius) tính toán một con số duy nhất đại diện cho tổng độ lớn của tất cả sai số đó.
        - Mục tiêu: Làm cho giá trị Loss này càng nhỏ càng tốt. Loss = 0 có nghĩa là XR bằng chính xác Y.

    b. Tính toán Gradient (g)
        - Ý nghĩa: Đây là bước quan trọng nhất. Phép toán này (lấy đạo hàm) giúp chúng ta tìm ra "hướng" mà Loss tăng nhanh nhất g chính là một ma trận cho biết mỗi phần tử trong R cần được thay đổi như thế nào để làm cho Loss giảm đi hiệu quả nhất.
        - Ví dụ tương tự: Hãy tưởng tượng bạn đang đứng trên một ngọn đồi và muốn đi xuống thung lũng (nơi có Loss thấp nhất). Gradient chính là hướng dốc lên nhất. Để đi xuống, bạn sẽ đi theo hướng ngược lại.

    c. Cập nhật R (Update)
        - Ý nghĩa: Chúng ta cập nhật lại ma trận R cũ để tạo ra một ma trận R mới tốt hơn.
        - Ta lấy R cũ trừ đi αg. Chúng ta dùng dấu trừ vì gradient g chỉ hướng "dốc lên", nên ta phải đi theo hướng ngược lại để "đi xuống".
        - α (alpha) là một con số nhỏ gọi là tốc độ học (learning rate). Nó quyết định xem mỗi lần cập nhật chúng ta sẽ bước đi một bước "nhỏ" hay "lớn".
