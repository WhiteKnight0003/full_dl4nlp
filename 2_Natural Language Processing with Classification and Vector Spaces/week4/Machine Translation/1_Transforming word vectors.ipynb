# 3. Gii php: Php bin i bng Ma trn (Transform with a Matrix)
- y l phn ct li ca s . N  xut mt gii php lgant: hc mt ma trn bin i (transformation matrix).
- Qu trnh:
    - Ly vector ca t gc trong ting Anh: [1, 0, 1] (vector ca "cat").
    - Nhn vector ny vi mt "ma trn bin i"  c "hc" trc .
    - Kt qu ca php nhn ny l mt vector mi: [2, 3, 2].

- Mc tiu: Vector mi [2, 3, 2] ny gi y nm trong khng gian vector ca ting Php. Mc tiu ca vic "hc" ma trn l  vector bin i ny gn nht c th vi vector thc t ca t dch ng.

- Trong v d ny, vector bin i [2, 3, 2] rt gn vi vector thc t ca t "chat" l [2, 3, 3]. iu ny cho thy php bin i  thnh cng. My tnh ch cn tm t c vector gn nht vi [2, 3, 2] trong kho t vng ting Php v n s tm thy "chat".

- X (Ma trn ngun): ng l ma trn cha thng tin ca cc t ting Anh. Chnh xc hn, n khng lu tr ch "cat", "dog" m lu tr vector (mt dy s) i din cho cc t .

- R (Ma trn bin i):  y chnh l "b no" m m hnh cn phi hc  thc hin vic nh x t khng gian ting Anh sang khng gian ting Php.

- Y  (Ma trn mc tiu): y l im cn lm r nht. Y  l ma trn mc tiu hay p n ng. N cha cc vector ca cc t ting Php tng ng m chng ta mun hng ti.

- ng vy, hnh nh ny m t chnh xc quy trnh thut ton  "gii" hay "tm" ra ma trn R.

- y l mt phng php rt ph bin trong lnh vc tr tu nhn to v hc my, c gi l H Gradient (Gradient Descent). Qu trnh ny din ra theo tng bc lp li  tinh chnh R ngy cng tt hn.
  
1 . Khi to R (Initialize R)
- u tin, chng ta khng bit R l g. V vy, ta bt u bng cch to ra mt ma trn R vi cc gi tr ngu nhin. y l "phng on" ban u ca chng ta.

2 . Vng lp (In a loop)
- Sau khi c R ban u, chng ta thc hin mt vng lp  lin tc ci thin n. Mi vng lp bao gm 3 bc sau:

    a. Tnh ton Sai s (Loss) 
        -  ngha: Cng thc ny o lng "mc  sai lch" gia kt qu d on (XR) v kt qu ng (Y).
        - XR - Y l ma trn cha cc sai s.
        - || ... ||_F (gi l chun Frobenius) tnh ton mt con s duy nht i din cho tng  ln ca tt c sai s .
        - Mc tiu: Lm cho gi tr Loss ny cng nh cng tt. Loss = 0 c ngha l XR bng chnh xc Y.

    b. Tnh ton Gradient (g)
        -  ngha: y l bc quan trng nht. Php ton ny (ly o hm) gip chng ta tm ra "hng" m Loss tng nhanh nht g chnh l mt ma trn cho bit mi phn t trong R cn c thay i nh th no  lm cho Loss gim i hiu qu nht.
        - V d tng t: Hy tng tng bn ang ng trn mt ngn i v mun i xung thung lng (ni c Loss thp nht). Gradient chnh l hng dc ln nht.  i xung, bn s i theo hng ngc li.

    c. Cp nht R (Update)
        -  ngha: Chng ta cp nht li ma trn R c  to ra mt ma trn R mi tt hn.
        - Ta ly R c tr i g. Chng ta dng du tr v gradient g ch hng "dc ln", nn ta phi i theo hng ngc li  "i xung".
        -  (alpha) l mt con s nh gi l tc  hc (learning rate). N quyt nh xem mi ln cp nht chng ta s bc i mt bc "nh" hay "ln".
