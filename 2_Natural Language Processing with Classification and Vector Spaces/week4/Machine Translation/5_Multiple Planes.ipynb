# Vai trò của việc tính Hash
- máy tính k làm việc với dấu +, - trực tiếp 
- Quy đổi dấu thành bit: Đầu tiên, các dấu +/- được quy đổi thành các bit 0 và 1 cho đơn giản.
    + dấu >= 0 (dương) → bit = 1
    + dấu < 0 (âm) → bit = 0

- Trong ví dụ, tổ hợp [+, +, -] trở thành dãy bit [1, 1, 0].

- Chuyển dãy bit thành số nguyên: Công thức hash = Σ (2^i × h_i) là cách tiêu chuẩn để chuyển một số nhị phân (dãy bit 0 và 1) thành một số thập phân.

- Trong ví dụ, dãy bit [h_1=1, h_2=1, h_3=0] được chuyển thành số 3.
