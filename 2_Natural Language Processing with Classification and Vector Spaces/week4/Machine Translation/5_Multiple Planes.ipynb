# Vai tr ca vic tnh Hash
- my tnh k lm vic vi du +, - trc tip 
- Quy i du thnh bit: u tin, cc du +/- c quy i thnh cc bit 0 v 1 cho n gin.
    + du >= 0 (dng)  bit = 1
    + du < 0 (m)  bit = 0

- Trong v d, t hp [+, +, -] tr thnh dy bit [1, 1, 0].

- Chuyn dy bit thnh s nguyn: Cng thc hash =  (2^i  h_i) l cch tiu chun  chuyn mt s nh phn (dy bit 0 v 1) thnh mt s thp phn.

- Trong v d, dy bit [h_1=1, h_2=1, h_3=0] c chuyn thnh s 3.
