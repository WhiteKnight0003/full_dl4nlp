

1 . Cp cu bn tri: T ging nhau, ngha khc nhau 
- "Where are you heading?" (Bn ang i u th?)
- "Where are you from?" (Bn n t u?)

- Hai cu ny c cu trc gn nh y ht, chia s vi nhau 4 trn 5 t. Mt m hnh ngn ng n gin da trn vic m t s thy chng rt ging nhau. Tuy nhin,  ngha ca chng li hon ton khc bit, thm ch l i lp.

2 . Cp cu bn phi: T khc nhau, ngha ging nhau 
- "What is your age?" (Tui ca bn l g?)
- "How old are you?" (Bn bao nhiu tui?)
- Hai cu ny khng c chung mt t vng no nhng li c cng mt  ngha. Mt m hnh n gin s khng th nhn ra mi lin h ny v coi chng l hai cu hon ton khc bit.

# Ti sao cn m hnh khng gian vector?
- Nhng v d trn cho thy vic ch da vo t ng l khng . Cc m hnh khng gian vector gii quyt vn  ny bng cch biu din mi t hoc cu di dng mt vector (mt chui cc con s) trong mt khng gian a chiu.
- Trong khng gian ny, khong cch v hng gia cc vector th hin mi quan h v mt ng ngha:
    + Cc vector ca hai cu "What is your age?" v "How old are you?" s nm rt gn nhau, cho thy chng c cng  ngha.
    + Cc vector ca hai cu "Where are you heading?" v "Where are you from?" s nm rt xa nhau hoc ch v cc hng khc nhau, cho thy  ngha ca chng khc bit.
  
- V vy, m hnh khng gian vector gip my tnh "hiu" c  ngha su xa hn thay v ch so khp b mt ca cc con ch.

# Cc ng dng chnh

1 . Trch xut thng tin (Information Extraction)
- y l qu trnh t ng rt ra nhng thng tin c cu trc (nh Ai, Ci g, Khi no,  u, Ti sao) t cc vn bn khng c cu trc.  Cc m hnh khng gian vector gip my tnh hiu c vai tr v mi quan h gia cc t trong cu, t  xc nh c u l ch th, hnh ng hay a im.

2 . Dch my (Machine Translation)
- y l ng dng dch t ng t ngn ng ny sang ngn ng khc.   dch chnh xc, my tnh khng ch dch tng t m phi hiu c c cu. V d, trong cu "You eat cereal from a bowl" (Bn n ng cc bng bt), m hnh vector s nm bt mi quan h gia "n", "ng cc" v "bt"  c th ti to li mt cu c  ngha tng ng trong ngn ng ch. 

3 . Chatbots
- Chatbots cn hiu c  nh ca ngi dng  a ra cu tr li ph hp.  Cc m hnh khng gian vector gip chatbots nhn ra s tng ng v khc bit v mt ng ngha. V d, trong cu "You buy something and someone else sells it" (Bn mua mt th g  v ngi khc bn n i), m hnh c th hiu rng "mua" (buy) v "bn" (sell) l hai mt ca cng mt giao dch, t  gip chatbot x l cc cuc hi thoi v thng mi mt cch thng minh hn.
