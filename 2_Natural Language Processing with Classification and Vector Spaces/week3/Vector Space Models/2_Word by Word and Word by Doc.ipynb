# 1 phương pháp cơ bản trong xử lý ngôn ngữ tự nhiên để hiểu mối quan hệ giữa các từ, được gọi là xây dựng ma trận đồng xuất hiện (co-occurrence matrix)


- Ý tưởng cốt lõi là :
- đếm số lần các từ xuất hiện cùng nhau trong một khoảng cách nhất định (k). Nguyên tắc đằng sau phương pháp này là "những từ xuất hiện trong ngữ cảnh giống nhau thường có ý nghĩa tương tự"

# Cách hoạt động qua ví dụ
- Hãy phân tích chi tiết cách tạo ra dòng của ma trận cho từ "data" trong ảnh:

1 . Xác định cửa sổ ngữ cảnh (k):
    + Tham số được chọn là k=2. Điều này có nghĩa là với mỗi từ mục tiêu, chúng ta sẽ nhìn vào 2 từ ở bên trái và 2 từ ở bên phải của nó

2 . Quét qua các câu:
    + Câu 1: "I like simple data"
        + Từ mục tiêu của chúng ta là "data".
        + Nhìn sang trái 2 từ, ta có cửa sổ ngữ cảnh là (like, simple).
        + Trong cửa sổ này, "data" xuất hiện cùng với "simple" (1 lần) và "like" (1 lần).

    + Câu 2: "I prefer simple raw data"
        + Từ mục tiêu vẫn là "data".
        + Nhìn sang trái 2 từ, ta có cửa sổ ngữ cảnh là (simple, raw).
        + Trong cửa sổ này, "data" xuất hiện cùng với "simple" (1 lần) và "raw" (1 lần).

3 . Tổng hợp kết quả:
    - Bây giờ, ta cộng số lần đếm được từ tất cả các câu:
        + simple: 1 lần (từ câu 1) + 1 lần (từ câu 2) = 2
        + raw: 0 lần (từ câu 1) + 1 lần (từ câu 2) = 1
        + like: 1 lần (từ câu 1) + 0 lần (từ câu 2) = 1
        + I: 0 lần (vì trong cả hai câu, "I" đều nằm ngoài khoảng cách k=2 so với "data").
  
- Kết quả này tạo ra dòng :
    [2, 1, 1, 0] cho từ "data" như trong hình. Bằng cách lặp lại quá trình này cho mọi từ trong từ vựng, chúng ta sẽ xây dựng được một ma trận hoàn chỉnh thể hiện mối quan hệ ngữ cảnh giữa các từ.
