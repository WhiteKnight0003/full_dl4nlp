# Cch hot ng qua v d
- Hy phn tch chi tit cch to ra dng ca ma trn cho t "data" trong nh:

1 . Xc nh ca s ng cnh (k):
    + Tham s c chn l k=2. iu ny c ngha l vi mi t mc tiu, chng ta s nhn vo 2 t  bn tri v 2 t  bn phi ca n

2 . Qut qua cc cu:
    + Cu 1: "I like simple data"
        + T mc tiu ca chng ta l "data".
        + Nhn sang tri 2 t, ta c ca s ng cnh l (like, simple).
        + Trong ca s ny, "data" xut hin cng vi "simple" (1 ln) v "like" (1 ln).

    + Cu 2: "I prefer simple raw data"
        + T mc tiu vn l "data".
        + Nhn sang tri 2 t, ta c ca s ng cnh l (simple, raw).
        + Trong ca s ny, "data" xut hin cng vi "simple" (1 ln) v "raw" (1 ln).

3 . Tng hp kt qu:
    - By gi, ta cng s ln m c t tt c cc cu:
        + simple: 1 ln (t cu 1) + 1 ln (t cu 2) = 2
        + raw: 0 ln (t cu 1) + 1 ln (t cu 2) = 1
        + like: 1 ln (t cu 1) + 0 ln (t cu 2) = 1
        + I: 0 ln (v trong c hai cu, "I" u nm ngoi khong cch k=2 so vi "data").
  
- Kt qu ny to ra dng :
    [2, 1, 1, 0] cho t "data" nh trong hnh. Bng cch lp li qu trnh ny cho mi t trong t vng, chng ta s xy dng c mt ma trn hon chnh th hin mi quan h ng cnh gia cc t.
