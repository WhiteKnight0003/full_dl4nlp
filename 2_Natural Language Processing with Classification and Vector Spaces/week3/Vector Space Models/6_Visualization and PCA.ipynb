# Cách PCA algothrm - Principal Component Analysis hoạt động

- Ý tưởng "chiếu bóng" , chúng ta chiếu lên một trục "thông minh" do PCA tính toán, chứ không phải trục X mặc định, để đảm bảo giữ lại được nhiều thông tin nhất có thể
  - Ban đầu (hình trên cùng bên trái): Dữ liệu của chúng ta nằm trên trục X và Y thông thường. Bạn có thể thấy các điểm dữ liệu tạo thành một đám mây hình elip, xiên lên trên.

  - PCA tìm trục mới (hình trên cùng bên phải): PCA sẽ không dùng trục X cũ. Thay vào đó, nó sẽ tìm một trục mới (gọi là thành phần chính thứ nhất - PC1) đi xuyên qua trung tâm của đám mây dữ liệu, theo cái hướng mà dữ liệu trải dài ra nhất. Trong hình, trục này đã được xoay lại để nằm ngang.

  - Chiếu bóng (hình dưới cùng): Sau khi đã xác định được trục mới tối ưu đó, bây giờ chúng ta mới thực hiện bước "chiếu bóng" mà bạn đã nói. Mỗi điểm dữ liệu sẽ được chiếu vuông góc xuống trục mới này.
