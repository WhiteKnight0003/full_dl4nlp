# Vocabulary - Từ vựng

- ✅ Tại sao ở Hình 1 lại phân tích từng tweet để lấy các từ duy nhất (tạo Vocabulary)? 

  - Vì đây là cách biểu diễn truyền thống nhất – gọi là Bag of Words (BoW):
  - Ý tưởng chính: Tạo một danh sách tất cả các từ duy nhất (unique tokens) xuất hiện trong tập văn bản, gọi là từ vựng 𝑉
  - Sau đó: Mỗi câu/tweet sẽ được biểu diễn thành vector nhị phân hoặc đếm số lần xuất hiện, theo thứ tự của các từ trong từ vựng.

- ✅ Vậy tại sao phải làm như vậy?
  
    1 . Đơn giản và dễ triển khai:
    - BoW là cách dễ nhất để chuyển văn bản thành số để máy học có thể xử lý.
    - Không cần học tham số, chỉ cần đếm.
    
    2 . Tạo đầu vào có kích thước cố định:
    - Dù câu dài hay ngắn, vector đặc trưng luôn có kích thước bằng V (số lượng từ trong từ vựng).
    - Phù hợp với các mô hình cổ điển như Logistic Regression, Naive Bayes, SVM.

- ❌ Tuy nhiên: Có những điểm yếu lớn
    - Không giữ ngữ cảnh (context): "I love cats" và "cats love I" → như nhau trong BoW.
    - Biểu diễn rời rạc (sparse): hầu hết các phần tử trong vector đều là 0.
    - Không hiểu quan hệ ngữ nghĩa giữa các từ: "happy" và "joyful" hoàn toàn tách biệt.
