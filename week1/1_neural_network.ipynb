```python
- Ví dụ ta có mô hình tự đoán giá nhà với :
    + input : size
    + ouput : price
    + ở giữa là Nơ-ron thực hiện một phép toán (thường là y = wx + b), Sau đó có thể áp dụng một hàm kích hoạt (activation function)
    + Dấu X màu đỏ: Dữ liệu thực tế (các căn nhà với diện tích và giá tương ứng)
    + Đường màu xanh: Đường dự đoán của mô hình (một đường thẳng)

- mô hình này sử dụng activation function là Relu - Rectified Linear Unit 
    - với Relu(x) = max(0, x) - tức min = 0 và max như kia
    -  ReLU giúp mô hình có phi tuyến tính, nhờ đó mô hình deep learning có thể học được các mối quan hệ phức tạp hơn tuyến tính.
```

- Bức hình này minh họa cách deep learning xử lý nhiều đặc trưng đầu vào (features) để dự đoán giá nhà. Đây là một mạng nơ-ron sâu cơ bản (fully connected neural network)

1. Đầu vào (Input layer):
- Gồm nhiều đặc trưng đầu vào:
    x₁: diện tích (size)
    x₂: số phòng ngủ (#bedrooms)
    x₃: mã vùng (zip code)
    x₄: độ giàu có (wealth)
➡️ Đây là vector đầu vào x = [x₁, x₂, x₃, x₄].

2. Hidden Layers (Tầng ẩn):
- Mỗi hình tròn là một nơ-ron nhân tạo.
- Các nơ-ron ở tầng ẩn nhận tín hiệu từ tất cả đầu vào.
- Chúng thực hiện phép tính:
        y = w₁x₁ + w₂x₂ + ... + wₙxₙ + b  (w1@x1 - tích vô hướng)
        và dùng hàm kích hoạt như ReLU để tạo phi tuyến

➡️ Các tầng này học cách trích xuất đặc trưng ẩn (latent features), ví dụ:
    "family size" , "walkability"  , "school quality" (được biểu diễn bên phải hình)


🎯 3. Output Layer (Tầng đầu ra):
Tầng cuối cùng trả về một giá trị duy nhất: giá nhà (price = y).

📦 Tổng kết:
| Thành phần    | Vai trò                                                 |
| ------------- | ------------------------------------------------------- |
| `x₁` → `x₄`   | Đầu vào mô hình                                         |
| Hidden layers | Học đặc trưng phức tạp                                  |
| Output        | Giá nhà dự đoán `y`                                     |
| `(x, y)`      | Một cặp dữ liệu đầu vào và đầu ra để huấn luyện mô hình |


🧠 Trọng số Wi là gì?
- Là thứ mô hình học được trong quá trình huấn luyện (training).
- Ban đầu: mô hình khởi tạo ngẫu nhiên
- Sau đó dùng Gradient Descent để điều chỉnh Wi sao cho mô hình dự đoán tốt hơn (giảm sai số giữa giá trị thực và dự đoán)

📌 Vị trí của Wi trong mô hình:
🔹 1. Các đường nối (connections) giữa nơ-ron
- Mỗi đường nối từ nơ-ron này sang nơ-ron khác có gắn một trọng số w
- Ví dụ: từ đầu vào x₁ (size) nối đến nơ-ron ẩn thứ nhất sẽ có trọng số w1,1
→ từ x₂ đến cùng nơ-ron đó là w2,1
