```python
- Vi du ta co m hinh t oan gia nha vi :
    + input : size
    + ouput : price
    +  gia la N-ron thc hin mt php ton (thng l y = wx + b), Sau  c th p dng mt hm kch hot (activation function)
    + Du X mu : D liu thc t (cc cn nh vi din tch v gi tng ng)
    + ng mu xanh: ng d on ca m hnh (mt ng thng)

- m hinh nay s dung activation function la Relu - Rectified Linear Unit 
    - vi Relu(x) = max(0, x) - tc min = 0 va max nh kia
    -  ReLU gip m hnh c phi tuyn tnh, nh  m hnh deep learning c th hc c cc mi quan h phc tp hn tuyn tnh.
```

- Bc hnh ny minh ha cch deep learning x l nhiu c trng u vo (features)  d on gi nh. y l mt mng n-ron su c bn (fully connected neural network)

1. u vo (Input layer):
- Gm nhiu c trng u vo:
    x: din tch (size)
    x: s phng ng (#bedrooms)
    x: m vng (zip code)
    x:  giu c (wealth)
 y l vector u vo x = [x, x, x, x].

2. Hidden Layers (Tng n):
- Mi hnh trn l mt n-ron nhn to.
- Cc n-ron  tng n nhn tn hiu t tt c u vo.
- Chng thc hin php tnh:
        y = wx + wx + ... + wx + b  (w1@x1 - tch v hng)
        v dng hm kch hot nh ReLU  to phi tuyn

 Cc tng ny hc cch trch xut c trng n (latent features), v d:
    "family size" , "walkability"  , "school quality" (c biu din bn phi hnh)


 3. Output Layer (Tng u ra):
Tng cui cng tr v mt gi tr duy nht: gi nh (price = y).

 Tng kt:
| Thnh phn    | Vai tr                                                 |
| ------------- | ------------------------------------------------------- |
| `x`  `x`   | u vo m hnh                                         |
| Hidden layers | Hc c trng phc tp                                  |
| Output        | Gi nh d on `y`                                     |
| `(x, y)`      | Mt cp d liu u vo v u ra  hun luyn m hnh |
