- Một Shallow Neural Network (Mạng Nơ-ron Nông) là một mạng nơ-ron nhân tạo chỉ có đúng một lớp ẩn (hidden layer).

# Để dễ hình dung, hãy xem nó như một bậc thang về độ phức tạp:

1 . Không có lớp ẩn (0 Hidden Layers):
    - Đây là các mô hình đơn giản nhất như Hồi quy Logistic (Logistic Regression).
    - Dữ liệu đi thẳng từ lớp đầu vào (Input Layer) đến lớp đầu ra (Output Layer).
    - Giống như hình 1  
    - 

2 . MỘT lớp ẩn (1 Hidden Layer):
    - Đây chính là Shallow Neural Network.
    - Cấu trúc: Lớp đầu vào → Lớp ẩn → Lớp đầu ra.
    - Lớp ẩn này cho phép mạng học được các đặc trưng và mối quan hệ phi tuyến tính (phức tạp hơn) từ dữ liệu.
    - Chính là hình 3 trong ảnh bạn gửi.
    - 

3 . NHIỀU lớp ẩn (2+ Hidden Layers):
    - Đây được gọi là Deep Neural Network (Mạng Nơ-ron Sâu - DNN).
    - Cấu trúc: Lớp đầu vào → Lớp ẩn 1 → Lớp ẩn 2 → ... → Lớp đầu ra.
    - Việc có nhiều lớp ẩn cho phép mạng học các đặc trưng ngày càng trừu tượng và phức tạp, phù hợp cho các bài toán khó như - nhận dạng hình ảnh, xử lý ngôn ngữ tự nhiên.

- -> Tóm lại: Thuật ngữ "Shallow" (nông) được dùng để phân biệt với "Deep" (sâu). Nếu mạng chỉ có một lớp ẩn, nó là "nông". Nếu có từ hai lớp ẩn trở lên, nó là "sâu".
