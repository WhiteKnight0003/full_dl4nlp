# Băm nhạy cục bộ (Locality Sensitive Hashing - LSH).

- Nguyên tắc chính LSH hoạt động dựa trên một ý tưởng đơn giản nhưng rất thông minh:
    - Các chấm tròn (dữ liệu): Tượng trưng cho các dữ liệu của bạn (ví dụ: các vector từ, hình ảnh...).
    - Các đường kẻ ("Planes"): Là những đường phân chia ngẫu nhiên. Mỗi đường kẻ đặt ra một câu hỏi đơn giản: "Điểm dữ liệu nằm ở bên này hay bên kia của đường kẻ?"
  
# nó "Nhạy cảm" với sự gần nhau (Locality Sensitive)
- Điểm đặc biệt của LSH nằm ở chỗ:
    - Những điểm gần nhau: Hãy nhìn vào nhóm các chấm trắng trong vòng tròn xám. Chúng ở rất gần nhau và tất cả chúng đều nằm về cùng một phía của cả hai đường kẻ. Do đó, chúng sẽ có cùng một mã băm (hash code).

    - Những điểm ở xa nhau: So sánh nhóm chấm trắng và nhóm chấm xanh. Chúng ở xa nhau và bị ít nhất một đường kẻ (đường nét đứt màu xanh) chia cắt. Vì nằm ở hai phía khác nhau của đường kẻ, chúng sẽ có mã băm khác nhau.


- ta có các planes (nó chính là các đường màu trắng hoặc màu xanh kia) 
- ví dụ trên đường màu xám nó giúp tôi nhóm 2 loại nút
  + ở phía trên nó giúp tôi nhóm nút màu trắng
  + ở phí dưới nó giúp tôi nhóm nút màu xanh


- ta sẽ vẽ đường P và chiếu các vector v1 , v2 lên đường P để xem các tập hợp điểm đó nằm bên này hay bên kia đường P


- Đoạn mã này định nghĩa một hàm để xác định một điểm (vector) nằm ở phía nào của một mặt phẳng. Đây chính là cách thực hiện "câu hỏi" mà chúng ta đã nói đến trong thuật toán Băm nhạy cục bộ (LSH).

- Hàm side_of_plane nhận vào 2 giá trị:
    - P: Vector đại diện cho mặt phẳng.
    - v: Vector đại diện cho điểm dữ liệu mà bạn muốn kiểm tra.
  
- Bước 1: Tính Tích vô hướng (Dot Product) 
    - Hành động: Dòng này thực hiện phép toán tích vô hướng giữa vector của mặt phẳng (P) và vector của điểm (v).
    - Tại sao: Trong hình học, tích vô hướng của hai vector cho chúng ta biết thông tin về góc giữa chúng.
  
- Bước 2: Lấy dấu của kết quả (Sign)
    - Hành động: Lấy dấu của kết quả dotproduct. Kết quả sẽ là:
        -  +1 nếu dotproduct là số dương.
        -  -1 nếu dotproduct là số âm.
        -  0 nếu dotproduct bằng 0.

    - Ý nghĩa: Dấu này cho biết điểm v nằm ở phía nào so với mặt phẳng P. Dấu +1 là một phía, dấu -1 là phía còn lại.
  
- Bước 3 và 4: Chuyển đổi và Trả về kết quả
    - Hành động: Chuyển đổi kết quả từ dạng dữ liệu của thư viện NumPy thành một con số Python đơn giản (scalar) và trả về con số đó.
