

1 . Cặp câu bên trái: Từ giống nhau, nghĩa khác nhau ↔️
- "Where are you heading?" (Bạn đang đi đâu thế?)
- "Where are you from?" (Bạn đến từ đâu?)

- Hai câu này có cấu trúc gần như y hệt, chia sẻ với nhau 4 trên 5 từ. Một mô hình ngôn ngữ đơn giản dựa trên việc đếm từ sẽ thấy chúng rất giống nhau. Tuy nhiên, ý nghĩa của chúng lại hoàn toàn khác biệt, thậm chí là đối lập.

2 . Cặp câu bên phải: Từ khác nhau, nghĩa giống nhau ✅
- "What is your age?" (Tuổi của bạn là gì?)
- "How old are you?" (Bạn bao nhiêu tuổi?)
- Hai câu này không có chung một từ vựng nào nhưng lại có cùng một ý nghĩa. Một mô hình đơn giản sẽ không thể nhận ra mối liên hệ này và coi chúng là hai câu hoàn toàn khác biệt.

# Tại sao cần mô hình không gian vector?
- Những ví dụ trên cho thấy việc chỉ dựa vào từ ngữ là không đủ. Các mô hình không gian vector giải quyết vấn đề này bằng cách biểu diễn mỗi từ hoặc câu dưới dạng một vector (một chuỗi các con số) trong một không gian đa chiều.
- Trong không gian này, khoảng cách và hướng giữa các vector thể hiện mối quan hệ về mặt ngữ nghĩa:
    + Các vector của hai câu "What is your age?" và "How old are you?" sẽ nằm rất gần nhau, cho thấy chúng có cùng ý nghĩa.
    + Các vector của hai câu "Where are you heading?" và "Where are you from?" sẽ nằm rất xa nhau hoặc chỉ về các hướng khác nhau, cho thấy ý nghĩa của chúng khác biệt.
  
- Vì vậy, mô hình không gian vector giúp máy tính "hiểu" được ý nghĩa sâu xa hơn thay vì chỉ so khớp bề mặt của các con chữ.

# Các ứng dụng chính

1 . Trích xuất thông tin (Information Extraction)
- Đây là quá trình tự động rút ra những thông tin có cấu trúc (như Ai, Cái gì, Khi nào, Ở đâu, Tại sao) từ các văn bản không có cấu trúc.  Các mô hình không gian vector giúp máy tính hiểu được vai trò và mối quan hệ giữa các từ trong câu, từ đó xác định được đâu là chủ thể, hành động hay địa điểm.

2 . Dịch máy (Machine Translation)
- Đây là ứng dụng dịch tự động từ ngôn ngữ này sang ngôn ngữ khác.  Để dịch chính xác, máy tính không chỉ dịch từng từ mà phải hiểu được cả câu. Ví dụ, trong câu "You eat cereal from a bowl" (Bạn ăn ngũ cốc bằng bát), mô hình vector sẽ nắm bắt mối quan hệ giữa "ăn", "ngũ cốc" và "bát" để có thể tái tạo lại một câu có ý nghĩa tương đương trong ngôn ngữ đích. 

3 . Chatbots
- Chatbots cần hiểu được ý định của người dùng để đưa ra câu trả lời phù hợp.  Các mô hình không gian vector giúp chatbots nhận ra sự tương đồng và khác biệt về mặt ngữ nghĩa. Ví dụ, trong câu "You buy something and someone else sells it" (Bạn mua một thứ gì đó và người khác bán nó đi), mô hình có thể hiểu rằng "mua" (buy) và "bán" (sell) là hai mặt của cùng một giao dịch, từ đó giúp chatbot xử lý các cuộc hội thoại về thương mại một cách thông minh hơn.
