- Computation Graph (đồ thị tính toán) là một cách biểu diễn chuỗi các phép toán được thực hiện trên dữ liệu đầu vào để ra được kết quả. Đây là một công cụ cực kỳ quan trọng trong deep learning vì nó là nền tảng để thực hiện backpropagation (lan truyền ngược) và tự động tính đạo hàm trong các framework như TensorFlow hoặc PyTorch.

- Đồ thị trong hình minh họa chính xác cách tính toán và thứ tự ưu tiên của phép toán.

- Cách tính toán với computation graph là thực hiện một chuỗi các phép toán đơn giản theo thứ tự định sẵn bởi các mũi tên, gọi là lan truyền tiến (forward propagation).

- Và có, nó có ưu tiên phép toán. Tuy nhiên, sự ưu tiên này không dựa vào quy tắc bạn phải nhớ (như "nhân chia trước, cộng trừ sau"), mà được quyết định bởi chính cấu trúc của đồ thị.

- Cách tính toán (Lan truyền tiến)
    - Bạn chỉ cần đi theo các mũi tên từ đầu vào đến đầu ra. Với ví dụ J=3(a+bc) và các giá trị a=5, b=3, c=2:
    - Tính nút u = bc:
        + Đồ thị cho thấy u cần đầu vào là b và c.
        + Tính toán: u = 3 * 2 = 6.
    - Tính nút v = a + u:   
        + Nút này cần đầu vào là a và kết quả từ nút u. Bạn không thể tính v trước khi có kết quả của u.
        + Tính toán: v = 5 + 6 = 11.
    - Tính nút cuối cùng J = 3v:
        + Nút này cần đầu vào là kết quả từ nút v.
        + Tính toán: J = 3 * 11 = 33.

👉 Kết quả cuối cùng là 33, khớp với những con số trên đồ thị.

Các mũi tên màu đỏ trong hình minh họa cho quá trình lan truyền ngược (backward pass). Bằng cách chia nhỏ phép tính, ta có thể dễ dàng áp dụng quy tắc chuỗi để tính đạo hàm của J theo từng biến a, b, c bằng cách đi ngược lại đồ thị. Đây là nền tảng cho việc huấn luyện các mô hình học máy phức tạp.
