
## 🧮 Logistic Regression - Giải thích chi tiết

### 📌 1. Mô hình Logistic Regression

Mục tiêu: Dự đoán xác suất một mẫu có nhãn y = 1 dựa vào vector đầu vào x.

ŷ = P(y = 1 | x )

#### Các thành phần:
- x: Vector đặc trưng đầu vào.
- w và b: Là các tham số (parameters) mà mô hình cần học. w là vector trọng số (weights) và b là một số gọi là thiên vị (bias).
- z = w^T@x + b: Là một phép tính tuyến tính, kết quả của nó có thể là một số thực bất kỳ (từ -∞ đến +∞). - @ là phép tích vô hướng 
- σ : Là hàm sigmoid, có nhiệm vụ biến đổi z thành một giá trị trong khoảng (0, 1).



### 🎯 2. Tại sao dùng hàm sigmoid?

Hàm sigmoid được dùng vì:
- Mô hình cần dự đoán xác suất ∈ [0, 1]
- Sigmoid có dạng:
    σ(z) = 1/(1+e^-z)


#### Tính chất:
- z → ∞ thì σ(z) → 1
- z → -∞ thì σ(z) → 0
- z = 0 thì σ(z) = 0.5

➡️ Phù hợp để phân biệt nhị phân.



### 🧠 3. Tại sao thêm x₀ = 1 và Tại sao θ₀ = b?
- Phương trình gốc: ŷ = σ(w^T x + b)
- Phương trình rút gọn: ŷ = σ(Θ^T x)

ta có z = (w1x1 + w2x2+...) + b
Để **gộp bias b vào vector trọng số**, ta mở rộng x thành x':

- x' = [1, x₁, x₂, ..., xₙ] ∈ R^(n_x + 1)
- θ = [b, w₁, w₂, ..., wₙ] ∈ R^(n_x + 1)

Khi đó:
z = wᵀx + b = θᵀ@x' (khi nhân tích vô hướng - nhân từng phần tử ở vị trí tương ứng với nhau)

✅ Giúp vector hóa toàn bộ mô hình.



### 📈 5. Đường cong sigmoid

- Dạng hình chữ "S"
- Cắt trục tung tại 0.5
- Dự đoán xác suất mượt mà, dễ phân loại:

| z | σ(z) |
|---|------|
| ≫ 0 | ≈ 1 |
| 0 | 0.5 |
| ≪ 0 | ≈ 0 |
