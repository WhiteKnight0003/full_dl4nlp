###  1. M hnh Logistic Regression

Mc tiu: D on xc sut mt mu c nhn y = 1 da vo vector u vo x.

y = P(y = 1 | x )

#### Cc thnh phn:
- x: Vector c trng u vo.
- w v b: L cc tham s (parameters) m m hnh cn hc. w l vector trng s (weights) v b l mt s gi l thin v (bias).
- z = w^T@x + b: L mt php tnh tuyn tnh, kt qu ca n c th l mt s thc bt k (t - n +). - @ l php tch v hng 
-  : L hm sigmoid, c nhim v bin i z thnh mt gi tr trong khong (0, 1).



###  2. Ti sao dng hm sigmoid?

Hm sigmoid c dng v:
- M hnh cn d on xc sut  [0, 1]
- Sigmoid c dng:
    (z) = 1/(1+e^-z)


#### Tnh cht:
- z   th (z)  1
- z  - th (z)  0
- z = 0 th (z) = 0.5

 Ph hp  phn bit nh phn.
