1 . Gradient Descent là gì?
- Gradient Descent (Giảm theo Gradient) là một thuật toán tối ưu hóa lặp đi lặp lại để tìm giá trị cực tiểu của một hàm số.

- Hãy tưởng tượng bạn đang đứng trên một sườn núi trong đêm tối và muốn đi xuống nơi thấp nhất (thung lũng). Bạn không thể nhìn xa, vì vậy tại mỗi bước chân, bạn sẽ tìm xem hướng nào dốc nhất xuống dưới và đi một bước nhỏ theo hướng đó. Bạn lặp lại quá trình này cho đến khi bạn không thể đi xuống được nữa, lúc đó bạn đã ở đáy. Gradient Descent hoạt động y hệt như vậy với hàm chi phí J(w,b).

- "Gradient" chính là một vector chỉ hướng dốc nhất tại một điểm.
- "Descent" (đi xuống) nghĩa là chúng ta di chuyển ngược hướng với gradient để đi đến điểm thấp hơn.

2 . Tại sao phải tìm w, b để J(w, b) nhỏ nhất?
- Hàm chi phí J(w, b) đại diện cho tổng sai số hay "mức độ sai" của mô hình trên toàn bộ tập dữ liệu.
- Một giá trị J(w, b) cao có nghĩa là các dự đoán của mô hình đang rất xa so với thực tế, tức là mô hình đang hoạt động tệ.
- Một giá trị J(w, b) thấp có nghĩa là các dự đoán của mô hình rất gần với thực tế, tức là mô hình đang hoạt động tốt.

- Do đó, mục tiêu của việc huấn luyện mô hình chính là tìm ra một cặp tham số w và b khiến cho tổng sai số J(w,b) là nhỏ nhất có thể (minimize). Khi đó, chúng ta có một mô hình dự đoán chính xác nhất.

3 . Tại sao J(w, b) phải là hàm lồi (Convex)?
- Hàm chi phí J(w, b) cần phải là hàm lồi (convex) để đảm bảo thuật toán Gradient Descent có thể tìm thấy 1 ĐIỂM CỰC TIỂU TOÀN CỤC DUY NHẤT một cách đáng tin cậy.

- Hàm lồi (Convex Function): Giống như một cái bát (như đồ thị 3D trên slide), nó chỉ có một điểm đáy duy nhất (cực tiểu toàn cục). Bất kể bạn bắt đầu từ đâu trên bề mặt, nếu bạn luôn đi xuống, bạn chắc chắn sẽ đến được điểm thấp nhất đó.

- Hàm không lồi (Non-convex Function): Giống như một dãy núi gập ghềnh với nhiều thung lũng (như đồ thị sóng ~ vẽ tay). Nó có nhiều "điểm đáy" khác nhau (cực tiểu cục bộ). Nếu bạn bắt đầu ở một vị trí không thuận lợi, Gradient Descent có thể đưa bạn xuống một thung lũng cạn và bị mắc kẹt ở đó, trong khi điểm đáy sâu nhất (lời giải tốt nhất) lại nằm ở một thung lũng khác.

- -> Bằng cách chọn một hàm mất mát (loss function) phù hợp để J(w,b) là hàm lồi, chúng ta đảm bảo rằng Gradient Descent sẽ không bị "lạc đường".

4 . Tại sao lại chọn một điểm rồi đi xuống đáy?
- Đây chính là cách thuật toán Gradient Descent hoạt động một cách trực quan:
    - Bước 1: Khởi tạo (Chọn một điểm ban đầu)
        + Chúng ta không biết w và b tốt nhất là gì khi bắt đầu. Vì vậy, thuật toán sẽ khởi tạo chúng với các giá trị ngẫu nhiên. Điểm (w, b) ngẫu nhiên này tương ứng với một điểm bất kỳ trên bề mặt của hàm chi phí J(w,b).

    - Bước 2: Lặp lại việc đi xuống
        + Tại điểm hiện tại, thuật toán tính gradient – tức là hướng dốc nhất đi lên.
        + Nó sẽ đi một bước nhỏ theo hướng ngược lại với gradient (hướng dốc nhất đi xuống). Kích thước của bước đi này được gọi là tốc độ học (learning rate).
        + Sau bước đi đó, chúng ta có một cặp (w, b) mới, tương ứng với một điểm thấp hơn trên bề mặt.
  
    - Bước 3: Dừng lại
        + Quá trình này được lặp đi lặp lại. Sau mỗi bước, chúng ta lại tiến gần hơn đến đáy. Thuật toán sẽ dừng lại khi nó đến rất gần đáy, nơi mà độ dốc gần như bằng 0 (không thể đi xuống thêm được nữa), hoặc sau một số lượng vòng lặp đã định trước.

    - -> Điểm cuối cùng mà thuật toán dừng lại chính là cặp (w, b) tối ưu mà chúng ta cần tìm.

- Giá trị x và y không phải là thứ chúng ta "giả sử" một cách ngẫu nhiên. Trong thực tế, x và y chính là dữ liệu mà bạn có.
- Nói một cách đơn giản:

- x là đầu vào (input): Đặc điểm mà bạn biết.
- y là đầu ra (output): Kết quả thực tế tương ứng với đầu vào đó.
- -> Mục tiêu của mô hình là học được mối quan hệ giữa x và y từ dữ liệu bạn cung cấp.

- sau vòng lặp 1 ta có w và b mới => thay lại vào hàm J(w, b) ban đầu sẽ tính được code sau từng bước 
- ví dụ ban đầu w = b = 0 - thay vào  J(w,b) = 16
- sau bước 1 , w = 1.6 , b = 0.8 thay lại vào J(w,b) = 0
