# 🧠 4. Trường hợp đặc biệt

| Bài toán                             | Loss Function                   | Ghi chú                                     |
| ------------------------------------ | ------------------------------- | ------------------------------------------- |
| **Autoencoder (Tái tạo)**            | `MSE` hoặc `MAE`                | Dự đoán lại chính đầu vào                   |
| **GAN (Generative Adversarial Net)** | `Binary Cross-Entropy`          | Phân biệt thật/giả                          |
| **Học tăng cường (RL)**              | `Policy Gradient`, `Q-Loss`     | Hàm mất mát tùy theo chiến lược/phần thưởng |
| **Ranking (Xếp hạng)**               | `Triplet Loss`, `Pairwise Loss` | So sánh cặp/nhóm để học thứ tự              |
