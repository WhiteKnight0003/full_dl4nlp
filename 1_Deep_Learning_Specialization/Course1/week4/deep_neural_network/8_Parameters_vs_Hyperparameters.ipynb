# Parameters (Tham s)
- y l nhng bin m m hnh t hc c t d liu trong qu trnh hun luyn. Mc tiu ca vic hun luyn mt m hnh hc su chnh l tm ra b tham s tt nht.
- V d chnh: Trng s W (weights) v  lch b (biases).

- Bn cht: Bn c th coi chng l "kin thc" ct li m mng n-ron  cht lc c t d liu. Cc gi tr ny c cp nht lin tc  mi vng lp ca thut ton ti u ha (nh Gradient Descent).

- -> Ni n gin, Parameters l nhng g m hnh hc.

# Hyperparameters (Siu tham s)
- y l nhng ci t, nhng cu hnh m bn (ngi pht trin) phi quyt nh trc khi qu trnh hun luyn bt u. Chng khng c hc t d liu m thay vo , chng iu khin cch thc qu trnh hc din ra.

    - Tc  hc (Learning rate ): Mc  iu chnh cc tham s W v b  mi bc.
    - S vng lp (Iterations): S ln thut ton s chy  cp nht tham s.
    - S lp n (L): Quyt nh  "su" ca mng n-ron.
    - S n v n (hidden units): Quyt nh  "rng" ca mi lp n.
    - La chn hm kch hot (Activation function): Dng hm ReLU, Sigmoid, hay Tanh,...
    - Cc siu tham s khc: Momentum, kch thc mini-batch, cc tham s iu chun (regularization),...

- -> Ni n gin, Hyperparameters l nhng g bn ci t  kim sot m hnh.
