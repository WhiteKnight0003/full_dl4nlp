🔷 1. Structured Data (Dữ liệu có cấu trúc)
- Dữ liệu có cấu trúc là dữ liệu dạng bảng (tabular data), thường thấy trong cơ sở dữ liệu hoặc file CSV, Excel, gồm các cột và hàng rõ ràng.

▸ Ví dụ 1: Dự đoán giá nhà
- Features (đặc trưng):
    Size (diện tích nhà): 2104 sqft
    #bedrooms (số phòng ngủ): 3
    Label (đầu ra cần dự đoán):
    Price: 400 (nghìn đô)

📌 Mục tiêu: Dùng các đặc trưng như diện tích, số phòng,... để dự đoán giá nhà.

🔶 2. Unstructured Data (Dữ liệu phi cấu trúc)
- Là dữ liệu không thể đưa vào bảng một cách trực tiếp như:
    Audio (âm thanh): Dự đoán nội dung từ tín hiệu giọng nói.
    Image (hình ảnh): Phân loại ảnh (ví dụ: mèo, chó,...).
    Text (văn bản): Dự đoán cảm xúc, phân loại văn bản, dịch ngôn ngữ,...

📌 Các dữ liệu này cần xử lý đặc biệt (như dùng CNN cho ảnh, RNN/Transformer cho văn bản/âm thanh).
