# Parameters (Tham số)
- Đây là những biến mà mô hình tự học được từ dữ liệu trong quá trình huấn luyện. Mục tiêu của việc huấn luyện một mô hình học sâu chính là tìm ra bộ tham số tốt nhất.
- Ví dụ chính: Trọng số W (weights) và độ lệch b (biases).

- Bản chất: Bạn có thể coi chúng là "kiến thức" cốt lõi mà mạng nơ-ron đã chắt lọc được từ dữ liệu. Các giá trị này được cập nhật liên tục ở mỗi vòng lặp của thuật toán tối ưu hóa (như Gradient Descent).

- -> Nói đơn giản, Parameters là những gì mô hình học.

# Hyperparameters (Siêu tham số)
- Đây là những cài đặt, những cấu hình mà bạn (người phát triển) phải quyết định trước khi quá trình huấn luyện bắt đầu. Chúng không được học từ dữ liệu mà thay vào đó, chúng điều khiển cách thức quá trình học diễn ra.

    - Tốc độ học (Learning rate α): Mức độ điều chỉnh các tham số W và b ở mỗi bước.
    - Số vòng lặp (Iterations): Số lần thuật toán sẽ chạy để cập nhật tham số.
    - Số lớp ẩn (L): Quyết định độ "sâu" của mạng nơ-ron.
    - Số đơn vị ẩn (hidden units): Quyết định độ "rộng" của mỗi lớp ẩn.
    - Lựa chọn hàm kích hoạt (Activation function): Dùng hàm ReLU, Sigmoid, hay Tanh,...
    - Các siêu tham số khác: Momentum, kích thước mini-batch, các tham số điều chuẩn (regularization),...

- -> Nói đơn giản, Hyperparameters là những gì bạn cài đặt để kiểm soát mô hình.
